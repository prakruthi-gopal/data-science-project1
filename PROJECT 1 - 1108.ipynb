```python
#Problem 15
q3 = """SELECT T.TransactionID, T.TransactionDate,T.UserID,T.ProductID,T.Quantity,U.User,U.Gender,U.Registered,U.Cancelled
        from users U LEFT OUTER JOIN transactions T ON T.UserID = U.UserID
 UNION
        SELECT T.TransactionID, T.TransactionDate,T.UserID,T.ProductID,T.Quantity,U.User,U.Gender,U.Registered,U.Cancelled  
        from transactions T LEFT OUTER JOIN users U ON T.UserID = U.UserID;"""
```

```python
#Problem 17
q5 = """SELECT DISTINCT users.UserID,transactions.ProductID
from users CROSS JOIN transactions ORDER BY users.UserID, transactions.ProductID;"""
```

```python
pysqldf(q5)
```

```python
#Problem 18
q6 = """SELECT users.UserID, products.ProductID, SUM(transactions.Quantity) as Quantity
from users INNER JOIN transactions ON transactions.UserID = users.UserID
INNER JOIN products ON products.ProductID=transactions.ProductID
GROUP BY users.UserID,transactions.ProductID;"""
```

```python
pysqldf(q6)
```

```python
#Problem 19
q7 = """SELECT DISTINCT x.TransactionID AS TransactionID_x, x.TransactionDate AS TransactionDate_x,
x.UserID AS UserID_x, x.ProductID AS ProductID_x, x.Quantity as Quantity_x,
y.TransactionID AS TransactionID_y, y.TransactionDate AS TransactionDate_y,
y.ProductID AS ProductID_y, y.Quantity as Quantity_y
from transactions x CROSS JOIN transactions y
ORDER BY x.TransactionID;"""
```

```python
pysqldf(q7)
```

```python
#Problem 20
q8 = """SELECT * 
from users LEFT OUTER JOIN transactions ON 
users.UserID = transactions.UserID
GROUP BY users.userID
HAVING MIN(transactions.TransactionDate) 
ORDER BY users.UserID;"""
```

```python
pysqldf(q8)
```

```python
#Problem 21
query = """SELECT users.UserID, users.User, users.Gender, users.Registered, users.Cancelled,
transactions.TransactionID, transactions.TransactionDate, transactions.ProductID, transactions.Quantity
from users, transactions;"""
data = pysqldf(query)
data
```
